# Load Dataset and Inspect Initial Shape

```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)
df = pd.read_csv("../datasets/raw/IndianWeatherRepository.csv")

print("Original Shape:", df.shape)
```

# Remove Redundant Features

```python
features_to_remove = [
    "country",
    "temperature_fahrenheit",
    "wind_mph",
    "pressure_in",
    "precip_in",
    "visibility_miles",
    "gust_mph"
]
df = df.drop(columns=features_to_remove)
print("Columns Removed Successfully")
```

# Verify Dataset Shape After Feature Removal

```python
print("New Shape:", df.shape)
```

# Validate Removed Columns

```python
removed_check = [
    col
    for col in features_to_remove
    if col not in df.columns
]
print("Removed Columns:")
print(removed_check)
```

# Save D3 Output Properly

```python
import os
os.makedirs("../datasets/processed", exist_ok=True)
df.to_csv(
    "../datasets/processed/D3_feature_removed.csv",
    index=False
)
print("D3 dataset saved successfully")
print("Location: ../datasets/processed/D3_feature_removed.csv")
```

# Verification
