# Imports

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)

df = pd.read_csv("../datasets/raw/IndianWeatherRepository.csv")

print("Dataset Loaded Successfully")
```

# Shape Analysis

```python
rows, cols = df.shape

print(f"Rows: {rows}")
print(f"Columns: {cols}")

# Insight
print("""
Insight:
The dataset contains 119,398 observations and 42 features,
making it sufficiently large for climate analytics,
machine learning, clustering, anomaly detection,
and risk prediction.
""")
```

# Datatype Analysis

```python
df.dtypes
```

# Datatype Summary

```python
datatype_summary = pd.DataFrame({
    "Column": df.columns,
    "Datatype": df.dtypes.values
})

datatype_summary
```

# Missing Value Analysis

```python
missing_values = df.isnull().sum()
missing_values
```

# Missing Value Report

```python
total_missing = missing_values.sum()

print("Total Missing Values:", total_missing)

# Insight
print("""
Insight:
No missing values were found.
No imputation is required.
Data quality is excellent.
""")
```

# Duplicate Analysis

```python
duplicates = df.duplicated().sum()

print("Duplicate Records:", duplicates)

# Insight

print("""
Insight:
No duplicate observations exist.
Dataset integrity is maintained.
""")
```

# Numerical vs Categorical Segmentation

```python
numerical_columns = df.select_dtypes(
    include=["int64", "float64"]
).columns

categorical_columns = df.select_dtypes(
    include=["object", "string"]
).columns

print("Numerical Features:", len(numerical_columns))
print("Categorical Features:", len(categorical_columns))
```

# Categorical Features

```python
pd.DataFrame({
    "Categorical Features": categorical_columns
})
```
