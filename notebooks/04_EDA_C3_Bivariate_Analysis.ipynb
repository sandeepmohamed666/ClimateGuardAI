# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_columns", None)
df = pd.read_csv("../datasets/raw/IndianWeatherRepository.csv")
print("Dataset Loaded Successfully")
```

# Temperature vs Humidity

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
   data=df,
   x="temperature_celsius",
   y="humidity",
   alpha=0.3
)
plt.title("Temperature vs Humidity")
plt.show()

# Insight

print("""
Investigates inverse relationship
between temperature and humidity.
Useful for heatwave analysis.
""")
```

# Temperature vs Pressure

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
   data=df,
   x="temperature_celsius",
   y="pressure_mb",
   alpha=0.3
)
plt.title("Temperature vs Pressure")
plt.show()
```

# Temperature vs Wind Speed

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
   data=df,
   x="temperature_celsius",
   y="wind_kph",
   alpha=0.3
)
plt.title("Temperature vs Wind Speed")
plt.show()
```

# PM2.5 vs PM10

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
   data=df,
   x="air_quality_PM2.5",
   y="air_quality_PM10",
   alpha=0.3
)
plt.title("PM2.5 vs PM10")
plt.show()

# Insight

print("""
Strong positive relationship expected.
Important for anomaly detection.
""")
```

# Carbon Monoxide vs PM2.5

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
   data=df,
   x="air_quality_Carbon_Monoxide",
   y="air_quality_PM2.5",
   alpha=0.3
)
plt.title("CO vs PM2.5")
plt.show()
```

# Rainfall vs Humidity

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
   data=df,
   x="humidity",
   y="precip_mm",
   alpha=0.3
)
plt.title("Humidity vs Rainfall")
plt.show()

# Insight

print("""
Humidity is expected to increase
before rainfall events.
Important predictor candidate.
""")
```

# Rainfall vs Cloud Cover

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
   data=df,
   x="cloud",
   y="precip_mm",
   alpha=0.3
)
plt.title("Cloud Cover vs Rainfall")
plt.show()
```

# Wind Speed vs Rainfall

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
   data=df,
   x="wind_kph",
   y="precip_mm",
   alpha=0.3
)
plt.title("Wind Speed vs Rainfall")
plt.show()
```

# Boxplot Temperature by Weather Condition

```python
plt.figure(figsize=(14,6))
sns.boxplot(
   data=df,
   x="condition_text",
   y="temperature_celsius"
)
plt.xticks(rotation=90)
plt.title("Temperature by Weather Condition")
plt.show()
```

# Rainfall by Weather Condition

```python
plt.figure(figsize=(14,6))
sns.boxplot(
   data=df,
   x="condition_text",
   y="precip_mm"
)
plt.xticks(rotation=90)
plt.title("Rainfall by Weather Condition")
plt.show()
```

# Region-wise Temperature

```python
top_regions = df["region"].value_counts().head(10).index
subset = df[
   df["region"].isin(top_regions)
]
plt.figure(figsize=(14,6))
sns.boxplot(
   data=subset,
   x="region",
   y="temperature_celsius"
)
plt.xticks(rotation=90)
plt.title("Temperature by Region")
plt.show()
```

# Correlation Preview

```python
correlation_pairs = [
   "temperature_celsius",
   "humidity",
   "pressure_mb",
   "precip_mm",
   "wind_kph",
   "air_quality_PM2.5",
   "air_quality_PM10"
]
df[correlation_pairs].corr()
```
